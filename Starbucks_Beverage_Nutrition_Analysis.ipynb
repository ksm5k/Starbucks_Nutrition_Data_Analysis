# Import Packages

```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns
```

# Create Dataframe With Starbucks Data

Read in the data using Pandas to create a dataframe.

```python
df = pd.read_csv('starbucks-menu-nutrition-drinks.csv')
```

Verify that the data was imported without issue by printing it. Printing allows us to see how the data is structured, too.

```python
print(df.head())
```

# Clean Data

Cells in this data set that are missing data are filled with '-'. For simplicity in this analysis, I will remove rows that are missing data.

```python
indexEmpty = df[ (df['Calories'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

```python
indexEmpty = df[ (df['Fat (g)'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

```python
indexEmpty = df[ (df['Carb. (g)'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

```python
indexEmpty = df[ (df['Fiber (g)'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

```python
indexEmpty = df[ (df['Protein'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

```python
indexEmpty = df[ (df['Sodium'] == '-')].index
df.drop(indexEmpty , inplace=True)
print(df)
```

It would be helpful to have a title for the column of drink names. I will call it "Drink". 

```python
df.rename(columns = {'Unnamed: 0': 'Drink'}, 
            inplace = True)
```

```python
print(df)
```

Check the types of data in each column. Convert numeric data to type 'float'. 

```python
print(df.dtypes)
```

```python
df['Calories'] = df['Calories'].astype('float')
df['Fat (g)'] = df['Fat (g)'].astype('float')
df['Carb. (g)'] = df['Carb. (g)'].astype('float')
df['Fiber (g)'] = df['Fiber (g)'].astype('float')
df['Protein'] = df['Protein'].astype('float')
df['Sodium'] = df['Sodium'].astype('float')
```

Now I will display a pie chart showing the percentage of drinks that fall within specified calorie ranges. 

```python
cals_above_300 = len(df[df["Calories"]>=300])
print(cals_above_300)
```

```python
cals_between_200_300 = len(df[(df["Calories"]< 300) & 
         (df["Calories"]>=200)])
print(cals_between_200_300)
```

```python
cals_between_100_200 = len(df[(df["Calories"] < 200) & 
         (df["Calories"]>=100)])
print(cals_between_100_200)
```

```python
cals_between_0_100 = len(df[(df["Calories"] < 100) & 
         (df["Calories"]>=0)])
print(cals_between_0_100)
```

Compare total number of drinks that we have counted with the number of drinks in the dataframe. It should be 92. 

```python
print(cals_above_300 + cals_between_200_300 + cals_between_100_200 + cals_between_0_100)
```

Create small list of counts of drinks that fall within certain calorie ranges, and a list of corresponding labels. These lists will be used to create the pie chart. 

```python
calories_data = [cals_above_300, cals_between_200_300, cals_between_100_200, cals_between_0_100]
```

```python
labels = ['Above 300', 'Between 200 and 300', 'Between 100 and 200', 'Less than 100']
```

```python
plt.pie(calories_data, labels = labels, autopct='%.0f%%')
plt.title('Percent of Drinks Within Various Calorie Ranges')
plt.show()
```
